# 00 â€” ARO Action Reference

Comprehensive reference for all ARO actions, organized by semantic role.
This notebook serves as a training reference and developer guide.

Actions are classified by **data flow direction**:

| Role | Direction | Description |
|------|-----------|-------------|
| **REQUEST** | External â†’ Internal | Pull data into the feature set |
| **OWN** | Internal â†’ Internal | Transform or decide within the feature set |
| **RESPONSE** | Internal â†’ External | Return a result or error to the caller |
| **EXPORT** | Internal â†’ Persistent/Event | Store, broadcast, or publish data |
| **FILE** | File system I/O | Read, write, or manipulate files/directories |
| **TERMINAL** | Terminal UI | Interactive terminal display |
| **SERVICE** | Infrastructure | Start/stop long-running services |
| **TEST** | Testing framework | Given/When/Then/Assert |

**Total actions: 60**

## REQUEST Actions

Pull data from an external source into the current feature set.

### Extract
**Verbs:** `extract`, `parse`, `get`  
**Prepositions:** `from`, `via`  
**Role:** REQUEST  

Extracts a named value from a structured source such as a request body, path parameters, query string, event payload, or any dictionary-like object. The qualifier (after `:`) specifies the field name when it differs from the result variable name.

```aro
(* Extract from HTTP path parameters *)
Extract the <id> from the <pathParameters: id>.

(* Extract from request body *)
Extract the <data> from the <request: body>.

(* Extract from event payload *)
Extract the <user> from the <event: user>.

(* Extract a nested field with alias *)
Extract the <email-address> from the <user: email>.
```

### Retrieve
**Verbs:** `retrieve`, `fetch`, `load`, `find`  
**Prepositions:** `from`  
**Role:** REQUEST  

Retrieves one or more records from an in-memory repository. Supports `where` clauses for filtering by field equality.

```aro
(* Retrieve all records *)
Retrieve the <users> from the <user-repository>.

(* Retrieve by ID *)
Retrieve the <user> from the <user-repository> where id = <user-id>.

(* Retrieve by field value *)
Retrieve the <orders> from the <order-repository> where status = "pending".
```

### Receive
**Verbs:** `receive`  
**Prepositions:** `from`, `via`  
**Role:** REQUEST  

Receives data from an external source such as a socket connection or incoming stream. Used in socket event handlers to get the message payload.

```aro
(* Receive message from socket packet *)
Receive the <message> from the <packet: buffer>.

(* Receive data via a channel *)
Receive the <frame> from the <socket: data>.
```

### Read
**Verbs:** `read`  
**Prepositions:** `from`  
**Role:** REQUEST  

Reads content from a file path. Supports format auto-detection: if the file is `.json`, `.yaml`, or `.csv` the result is a structured object/list. Plain text files return a String.

```aro
(* Read plain text *)
Read the <content> from the <config-file>.

(* Read JSON â€” result is a dict *)
Read the <settings> from "config.json".

(* Read CSV â€” result is a list of dicts *)
Read the <records> from "data.csv".
```

### Request
**Verbs:** `request`, `http`  
**Prepositions:** `from`, `to`, `via`, `with`  
**Role:** REQUEST  

Makes an HTTP request to an external URL. Default method is GET. Use `to` for POST/PUT with a body.

```aro
(* GET request *)
Request the <forecast> from the <weather-api-url>.

(* POST with body *)
Request the <response> to the <endpoint> with <payload>.

(* With qualifier for method override *)
Request the <result: POST> from the <api-url> with <data>.
```

### List
**Verbs:** `list`  
**Prepositions:** `from`  
**Role:** REQUEST  

Lists the contents of a directory, returning a list of file/directory names.

```aro
(* List directory entries *)
List the <entries> from the <uploads-directory>.

(* List with path string *)
List the <files> from "/var/data/reports".
```

### Stat
**Verbs:** `stat`  
**Prepositions:** `for`  
**Role:** REQUEST  

Returns file or directory metadata: `name`, `path`, `size`, `isDirectory`, `createdAt`, `modifiedAt`, `extension`.

```aro
(* Get file metadata *)
Stat the <info> for the <document-path>.
Extract the <size> from the <info: size>.
Extract the <modified> from the <info: modifiedAt>.
```

### Exists
**Verbs:** `exists`  
**Prepositions:** `for`  
**Role:** REQUEST  

Checks whether a file or directory exists. Returns a Boolean.

```aro
(* Check file existence *)
Exists the <found> for the <backup-file>.

(* Use in conditional *)
Exists the <cache-exists> for the <cache-path>.
Compute the <result> from <data> when <cache-exists>.
```

### Prompt
**Verbs:** `prompt`, `ask`  
**Prepositions:** `with`, `from`  
**Role:** REQUEST  

Displays a prompt to the user in the terminal and reads their text input. Used in interactive CLI applications.

```aro
(* Prompt for input *)
Prompt the <username> with "Enter your username: ".
Prompt the <password> with "Password: ".
```

### Select
**Verbs:** `select`, `choose`  
**Prepositions:** `from`, `with`  
**Role:** REQUEST  

Displays a numbered menu of options in the terminal and reads the user's selection. Returns the selected item.

```aro
(* Show menu and get choice *)
Select the <action> from the <menu-options>.
Select the <region> from <regions> with "Choose a region:".
```

### Stream
**Verbs:** `stream`, `subscribe`  
**Prepositions:** `from`, `with`  
**Role:** REQUEST  

Opens a lazy stream from an event source or data producer. Used with ARO-0051 streaming execution for large datasets.

```aro
(* Open a stream *)
Stream the <log-events> from the <event-source>.
Stream the <rows> from the <large-csv-file>.
```

## OWN Actions

Transform data or make decisions entirely within the feature set. No external I/O.

### Compute
**Verbs:** `compute`, `calculate`, `derive`  
**Prepositions:** `from`, `for`, `with`  
**Role:** OWN  

Performs arithmetic, string operations, or built-in transformations. Supports `+`, `-`, `*`, `/`, `%` operators and qualifiers like `length`, `uppercase`, `lowercase`, `hash`, `count`, `intersect`, `difference`, `union`.

```aro
(* Arithmetic *)
Compute the <total> from <price> * <quantity>.
Compute the <discount> from <total> * 0.1.
Compute the <final-price> from <total> - <discount>.

(* String operations *)
Compute the <upper: uppercase> from the <name>.
Compute the <len: length> from the <message>.

(* Set operations *)
Compute the <shared: intersect> from <set-a> with <set-b>.

(* Boolean from comparison â€” useful as when-guard *)
Compute the <is-admin> from <role> = "admin".
```

### Validate
**Verbs:** `validate`, `verify`, `check`  
**Prepositions:** `for`, `against`, `with`  
**Role:** OWN  

Validates a value against a schema, pattern, or rule. Returns a Boolean. Common qualifiers: `email`, `url`, `uuid`, `range`, `required`.

```aro
(* Validate email *)
Validate the <valid-email> for the <email-address>.

(* Validate against regex pattern *)
Validate the <valid-zip> for the <zip-code> against /^\d{5}$/.

(* Validate required fields *)
Validate the <has-name> for the <user> with <name-schema>.
```

### Compare
**Verbs:** `compare`, `match`  
**Prepositions:** `against`, `with`, `to`  
**Role:** OWN  

Compares two values and binds a Boolean result. Used for equality, ordering, and pattern matching.

```aro
(* Equality comparison *)
Compare the <is-match> against the <expected-hash> with <actual-hash>.

(* Use result as conditional *)
Compare the <passwords-match> against <hash> with <input-hash>.
Return a <Forbidden: status> for the <login> when not <passwords-match>.
```

### Transform
**Verbs:** `transform`, `convert`, `map`  
**Prepositions:** `from`, `into`, `to`  
**Role:** OWN  

Converts a value from one type or format to another. Qualifiers specify the target type: `int`, `string`, `float`, `bool`, `date`, `json`, `markdown`.

```aro
(* Type conversion *)
Transform the <count: int> from the <count-string>.
Transform the <price: float> from the <price-text>.

(* Format conversion *)
Transform the <json-body: json> from the <response-text>.
Transform the <html-output: markdown> from the <markdown-text>.
```

### Create
**Verbs:** `create`, `build`, `construct`  
**Prepositions:** `with`, `from`, `for`, `to`  
**Role:** OWN  

Creates a new in-memory object or data structure from a dict literal or existing variables. Does NOT persist â€” use `Store` to save.

```aro
(* Create object from dict literal *)
Create the <user> with { name: <name>, email: <email>, role: "user" }.

(* Create with defaults *)
Create the <counter> with { key: <id>, count: 0 }.

(* Create list *)
Create the <empty-list> with [].
```

### Update
**Verbs:** `update`, `modify`, `change`, `set`, `configure`  
**Prepositions:** `with`, `to`, `for`, `from`  
**Role:** OWN  

Produces a new object with one or more fields changed. The original is immutable â€” `Update` always returns a new binding.

```aro
(* Update a single field *)
Update the <active-user> with { status: "active" } from the <user>.

(* Configure runtime settings *)
Configure the <timeout-setting: timeout> with 30.
Configure the <cache: maxSize> with 1000.
```

### Sort
**Verbs:** `sort`, `order`, `arrange`  
**Prepositions:** `for`, `with`  
**Role:** OWN  

Sorts a collection. Default is ascending order. Use qualifier `desc` for descending, or specify a field name.

```aro
(* Sort ascending *)
Sort the <sorted-users> for the <users>.

(* Sort descending *)
Sort the <newest-first: desc> for the <messages>.

(* Sort by field *)
Sort the <by-name: name> for the <products>.
```

### Merge
**Verbs:** `merge`, `combine`  
**Prepositions:** `with`, `from`  
**Role:** OWN  

Merges two collections or objects. For dicts, later keys overwrite earlier ones. For lists, concatenates.

```aro
(* Merge two lists *)
Merge the <all-users> with <active-users> from <archived-users>.

(* Merge two dicts (second overrides first) *)
Merge the <full-config> with <defaults> from <overrides>.
```

### Delete
**Verbs:** `delete`, `remove`, `destroy`, `clear`  
**Prepositions:** `from`, `for`  
**Role:** OWN  

Removes an element from a collection or repository. When used with a repository, deletes the persisted record.

```aro
(* Remove from in-memory list *)
Delete the <updated-list> from the <item> in the <item-list>.

(* Delete from repository *)
Delete the <removed> from the <user-repository> where id = <user-id>.
```

### Map
**Verbs:** `map`  
**Prepositions:** `from`, `to`  
**Role:** OWN  

Transforms each element in a collection using a qualifier or field accessor.

```aro
(* Extract a field from each item *)
Map the <names> from the <users: name>.

(* Apply transformation to each element *)
Map the <prices: float> from the <price-strings>.
```

### Reduce
**Verbs:** `reduce`, `aggregate`  
**Prepositions:** `from`, `with`  
**Role:** OWN  

Aggregates a collection into a single value. Built-in qualifiers: `sum`, `avg`, `count`, `min`, `max`, `first`, `last`.

```aro
(* Sum a numeric field *)
Reduce the <total-price: sum> from the <line-items: price>.

(* Count items *)
Reduce the <count: count> from the <orders>.

(* Average *)
Reduce the <average-score: avg> from the <scores>.
```

### Filter
**Verbs:** `filter`  
**Prepositions:** `from`  
**Role:** OWN  

Filters a collection by a predicate. Uses `where` clauses for field-value conditions.

```aro
(* Filter by field value *)
Filter the <active-users> from the <users> where status = "active".

(* Filter by numeric comparison *)
Filter the <adults> from the <applicants> where age >= 18.
```

### Call
**Verbs:** `call`, `invoke`  
**Prepositions:** `from`, `to`, `with`, `via`  
**Role:** OWN  

Invokes an external service method or plugin action. Used for inter-service communication and plugin execution.

```aro
(* Call external service *)
Call the <result> from the <payment-service> with <charge-data>.

(* Call plugin action *)
Call the <html: Markdown.ToHTML> from the <markdown-text>.
```

### Execute
**Verbs:** `execute`, `exec`, `run`, `shell`  
**Prepositions:** `on`, `with`, `for`  
**Role:** OWN  

Executes a shell command and captures its stdout. The result is a String.

```aro
(* Run a shell command *)
Execute the <output> with "git log --oneline -5".

(* Run with arguments from a variable *)
Execute the <result> with <shell-command>.
```

### Parse
**Verbs:** `parse` (HTML/XML variant)  
**Prepositions:** `from`  
**Role:** OWN  

Parses HTML or XML content into a structured document. Supports extractors: `links`, `text`, `title`, `headings`. Also parses RFC 8288 Link headers.

```aro
(* Parse HTML and extract links *)
Parse the <document> from the <html-content>.
Extract the <links> from the <document: links>.

(* Parse Link header *)
Parse the <pagination> from the <link-header>.
Extract the <next-url> from the <pagination: next>.
```

### Accept
**Verbs:** `accept`  
**Prepositions:** `on`  
**Role:** OWN  

Validates and applies a state transition on an entity in a repository. Emits a `StateTransition` event on success. Rejects invalid transitions silently.

```aro
(* Transition order to 'shipped' *)
Accept the <order> on the <placed-state>.

(* Transition user to 'suspended' *)
Accept the <account> on the <active-state>.
```

### Join
**Verbs:** `join`  
**Prepositions:** `from`  
**Role:** OWN  

Joins the elements of a list into a single string. The qualifier specifies the separator.

```aro
(* Join with comma *)
Join the <csv-line: ,> from the <values>.

(* Join with space *)
Join the <sentence: " "> from the <words>.

(* Join with newline *)
Join the <report: \n> from the <lines>.
```

### Split
**Verbs:** `split`  
**Prepositions:** `from`  
**Role:** OWN  

Splits a string by a regex delimiter into a list of strings. The delimiter is a regex literal `/pattern/`.

```aro
(* Split by comma *)
Split the <fields> from the <csv-line> by /,/.

(* Split by whitespace *)
Split the <words> from the <sentence> by /\s+/.

(* Split by newline *)
Split the <lines> from the <content> by /\n/.
```

### Sleep
**Verbs:** `sleep`, `delay`, `pause`  
**Prepositions:** `for`, `with`  
**Role:** OWN  

Suspends execution of the current feature set for a given number of seconds. Non-blocking â€” other feature sets continue running.

```aro
(* Sleep for 2 seconds *)
Sleep the <pause> for 2.

(* Sleep with variable duration *)
Sleep the <wait> for <retry-delay>.
```

## RESPONSE Actions

Send a result or error back to the caller and end the feature set's execution path.

### Return
**Verbs:** `return`, `respond`  
**Prepositions:** `for`, `to`, `with`  
**Role:** RESPONSE  

Returns a response from the current feature set. For HTTP feature sets, this sets the HTTP response status and body. The qualifier on the result names the status code.

```aro
(* Return success with body *)
Return an <OK: status> with <user>.

(* Return created *)
Return a <Created: status> with <new-record>.

(* Return no-content *)
Return a <NoContent: status> for the <deletion>.

(* Return in non-HTTP context *)
Return an <OK: status> for the <startup>.
```

### Throw
**Verbs:** `throw`, `raise`, `fail`  
**Prepositions:** `for`  
**Role:** RESPONSE  

Throws an error that propagates to the caller. For HTTP, this produces an error response. The qualifier names the error type.

```aro
(* Throw not found *)
Throw a <NotFound: error> for the <user>.

(* Throw validation error *)
Throw a <BadRequest: error> for the <invalid-input>.

(* Throw with guard *)
Throw a <Unauthorized: error> for the <request> when not <is-authenticated>.
```

## EXPORT Actions

Persist, broadcast, or publish data outside the current feature set boundary.

### Store
**Verbs:** `store`, `save`, `persist`  
**Prepositions:** `into`, `to`, `in`  
**Role:** EXPORT  

Persists an object into an in-memory repository. Automatically assigns a UUID `id` if none is present. Use the qualifier to name the stored entity type.

```aro
(* Store with type annotation *)
Store the <stored-user: user> into the <user-repository>.

(* Update existing record *)
Store the <updated-order> into the <order-repository>.
```

### Publish
**Verbs:** `publish`, `export`, `expose`, `share`  
**Prepositions:** `with` (syntax: `Publish as <alias> <variable>`)  
**Role:** EXPORT  

Publishes a variable so it is accessible to other feature sets within the same business activity scope.

```aro
(* Publish config for use by event handlers *)
Publish as <app-config> <config>.

(* Publish server handle *)
Publish as <http-server> <server>.
```

### Log
**Verbs:** `log`, `print`, `output`, `debug`  
**Prepositions:** `for`, `to`, `with`  
**Role:** EXPORT  

Writes a message to the console or a log file. Use string literals for fixed messages, or variable references for dynamic values.

```aro
(* Log string literal *)
Log "Server starting on port 8080" to the <console>.

(* Log variable *)
Log <error-message> to the <console>.

(* Log with prefix *)
Log "User created:" to the <console>.
Log <user-id> to the <console>.
```

### Send
**Verbs:** `send`, `dispatch`  
**Prepositions:** `to`, `via`, `with`  
**Role:** EXPORT  

Sends a message to a specific socket connection. Used in socket event handlers.

```aro
(* Send to a specific connection *)
Send the <welcome-message> to the <connection>.

(* Send with extra data *)
Send the <response> to the <client-connection> with <metadata>.
```

### Emit
**Verbs:** `emit`  
**Prepositions:** `with`, `to`  
**Role:** EXPORT  

Emits a domain event on the event bus. Other feature sets with matching handlers will be triggered asynchronously.

```aro
(* Emit with payload *)
Emit a <UserCreated: event> with <user>.

(* Emit with explicit variable *)
Emit an <OrderShipped: event> with <order>.

(* Emit state change *)
Emit a <PaymentProcessed: event> with <payment>.
```

### Notify
**Verbs:** `notify`, `alert`, `signal`  
**Prepositions:** `to`, `for`, `with`  
**Role:** EXPORT  

Sends a notification to a recipient object or collection. Feature sets named `{Name}: NotificationSent Handler` with optional `when` guards receive the notification.

```aro
(* Notify single recipient *)
Notify the <recipient> with "Your order has shipped.".

(* Notify all in a list *)
Notify the <admin-users> with "System maintenance in 5 minutes.".

(* Handler that filters by recipient field *)
(* (Admin Email: NotificationSent Handler) when <role> = "admin" { ... } *)
```

### Write
**Verbs:** `write`  
**Prepositions:** `to`, `into`  
**Role:** EXPORT  

Writes content to a file. If the value is a dict or list it is serialized to JSON; strings are written as-is.

```aro
(* Write string to file *)
Write the <report> to the <output-file>.

(* Write dict as JSON *)
Write the <config-data> to "settings.json".
```

### Append
**Verbs:** `append`  
**Prepositions:** `to`, `into`  
**Role:** EXPORT  

Appends content to an existing file. Creates the file if it does not exist.

```aro
(* Append log entry *)
Append the <log-entry> to the <audit-log>.

(* Append with newline *)
Append the <line> to "application.log".
```

### Schedule
**Verbs:** `schedule`  
**Prepositions:** `with`  
**Role:** EXPORT  

Schedules a recurring domain event to fire every N seconds. Other feature sets can listen to the scheduled event by name.

```aro
(* Fire an event every 30 seconds *)
Schedule the <heartbeat> with 30.

(* Handler that receives the event *)
(* (Health Check: heartbeat Handler) { ... } *)
```

## FILE Actions

File system operations beyond read/write.

### Make
**Verbs:** `make`, `touch`, `mkdir`  
**Prepositions:** `to`, `for`, `at`  
**Role:** OWN  

Creates a new empty file or directory. Use qualifier `directory` or `file` to disambiguate.

```aro
(* Create directory *)
Make the <uploads: directory> at the <uploads-path>.

(* Create empty file *)
Make the <lock-file: file> at "/tmp/app.lock".
```

### Copy
**Verbs:** `copy`  
**Prepositions:** `to`  
**Role:** OWN  

Copies a file or directory to a destination path.

```aro
(* Copy file *)
Copy the <backup> to the <backup-destination>.

(* Copy directory *)
Copy the <snapshot> to "/backups/snapshot-2025".
```

### Move
**Verbs:** `move`, `rename`  
**Prepositions:** `to`  
**Role:** OWN  

Moves or renames a file or directory.

```aro
(* Move file *)
Move the <processed-file> to the <archive-path>.

(* Rename *)
Move the <old-config> to "config.backup.yaml".
```

## TERMINAL Actions

Interactive terminal UI â€” render sections, clear screen, show panels.

### Clear
**Verbs:** `clear`  
**Prepositions:** `for`  
**Role:** TERMINAL  

Clears the terminal screen and resets the section compositor. No-op in non-TTY mode.

```aro
(* Clear at startup *)
Clear the <display> for the <terminal>.
```

### Show
**Verbs:** `show`  
**Prepositions:** `for`  
**Role:** TERMINAL  

Shows a value or section in the terminal without using the section compositor.

```aro
(* Show content *)
Show the <help-text> for the <terminal>.
```

### Render
**Verbs:** `render`  
**Prepositions:** `to`  
**Role:** TERMINAL  

Renders a template into a named terminal section. Uses the section compositor: first render places the section, subsequent renders diff and update in-place without clearing.

```aro
(* Render template to terminal *)
Render the <screen> to the <terminal>.

(* Re-render same section (updates in-place) *)
Render the <status-bar> to the <terminal>.
```

### Repaint
**Verbs:** `repaint`, `patch`  
**Prepositions:** `at`, `to`  
**Role:** TERMINAL  

Updates a specific section of the terminal display, used to partially refresh the UI without redrawing everything.

```aro
(* Repaint a named section *)
Repaint the <header> at the <terminal>.
Repaint the <status-line> at the <terminal>.
```

## SERVICE Actions

Manage long-running services and infrastructure.

### Start
**Verbs:** `start`  
**Prepositions:** `with`  
**Role:** SERVICE  

Starts a service. Common targets: `http-server` (requires `contract`), `file-monitor`, `socket-server`.

```aro
(* Start HTTP server *)
Start the <http-server> with <contract>.

(* Start file monitor *)
Start the <file-monitor> with ".".

(* Start socket server *)
Start the <socket-server> with <socket-config>.
```

### Stop
**Verbs:** `stop`  
**Prepositions:** `with`  
**Role:** SERVICE  

Stops a running service gracefully.

```aro
(* Stop HTTP server on shutdown *)
Stop the <http-server> with <application>.
```

### Listen
**Verbs:** `listen`, `await`  
**Prepositions:** `on`, `for`, `to`  
**Role:** SERVICE  

Registers a listener on an event channel or port.

```aro
(* Listen on a channel *)
Listen the <incoming> on the <event-channel>.
```

### Keepalive
**Verbs:** `keepalive`, `wait`, `block`  
**Prepositions:** `for`  
**Role:** SERVICE  

Blocks the current feature set (typically `Application-Start`) until a shutdown signal is received (SIGINT/SIGTERM). Required for server and file-watcher applications to stay alive.

```aro
(* Keep application running *)
Keepalive the <application> for the <events>.
```

### Connect
**Verbs:** `connect`  
**Prepositions:** `to`, `with`  
**Role:** SERVICE  

Establishes a TCP socket connection to a remote host.

```aro
(* Connect to server *)
Connect the <connection> to the <server-address>.

(* Connect with config *)
Connect the <socket> to "localhost:8080".
```

### Broadcast
**Verbs:** `broadcast`  
**Prepositions:** `to`, `via`  
**Role:** SERVICE  

Broadcasts a message to all currently connected socket clients.

```aro
(* Broadcast to all connected clients *)
Broadcast the <notification> to the <socket-server>.

(* Broadcast update *)
Broadcast the <price-update> to the <ws-server>.
```

### Close
**Verbs:** `close`, `disconnect`, `terminate`  
**Prepositions:** `with`, `from`  
**Role:** SERVICE  

Closes a socket connection gracefully.

```aro
(* Close connection *)
Close the <session> with <connection>.
```

## TEST Actions

BDD-style testing framework. Co-located with feature set files.

### Given
**Verbs:** `given`  
**Prepositions:** `with`  
**Role:** TEST  

Sets up the test context with initial state.

```aro
(* Set up test data *)
Given the <context> with <test-user>.
```

### When
**Verbs:** `when`  
**Prepositions:** `from`  
**Role:** TEST  

Executes the action under test.

```aro
(* Execute the feature under test *)
When the <response> from the <createUser>.
```

### Then
**Verbs:** `then`  
**Prepositions:** `with`  
**Role:** TEST  

Verifies the result against expected output.

```aro
(* Verify result *)
Then the <created-user> with <expected-user>.
```

### Assert
**Verbs:** `assert`  
**Prepositions:** `for`, `with`  
**Role:** TEST  

Asserts that a value matches an expectation. Fails the test with a clear message if not.

```aro
(* Assert equality *)
Assert the <valid-email> for the <email-result>.

(* Assert with expected value *)
Assert the <status-ok> for the <response> with "200".
```
